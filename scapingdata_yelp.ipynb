```python
def GetRes(cuisine,location):
    url= "https://api.yelp.com/v3/businesses/search"
    key="cjClr8c3q0GVoc9Yiv-DpVQcZ6pu6P1rSwBMBKlrqALd7xnZ6i682GhTXJfDtYzVuFZSggLB-ls0e7b9GNFmrAX6LTdWEPAKuug6Rn1VhUYaLaNREDixTzRDTwAQYnYx"
    headers={
        "Authorization":'Bearer %s' % key
        }
    total_businesses=[]
    businesses=[]
    count=20
    for c in cuisine:
        print("one cuisine started")
        for cnt in range(count):
            offset=50*cnt
            parameters={
                "location": location,
                "term":c+" "+"restaurants",
                "limit":50,
                "offset":offset
               }
            response=requests.get(url,data="",headers=headers, params=parameters)
            if response.status_code==200: #success
                data=response.json()['businesses']
                for d in data:
                    if d['id'] not in total_businesses:
                        dict = {}
                        dict['insertedAtTimestamp'] = str(datetime.datetime.now())
                        dict['business ID'] = d['id']
                        dict['name'] = d['name']
                        dict['cuisine'] = c
                        dict['address'] = " ".join(d['location']['display_address'])
                        dict['rating'] = int(d['rating'])
                        dict['zip_code'] = d['location']["zip_code"]
                        dict['latitude'] = str(d['coordinates']['latitude'])
                        dict['longitude'] = str(d['coordinates']['longitude'])
                        dict['review_number'] = int(d['review_count'])

                        businesses.append(dict)
                        total_businesses.append(d['id'])

            elif response.status_code==400:
                continue

    return businesses
    
    
    
            
```
